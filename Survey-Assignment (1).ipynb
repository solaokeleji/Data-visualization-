```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/FinalModule_Coursera/data/Topic_Survey_Assignment.csv"
df_survey = pd.read_csv(path, index_col=0)
df_survey.head()
```

```python
df_survey.sort_values(by= 'Very interested', ascending=False, axis=0, inplace=True)
df_survey.head()
```

```python
#Convert the numbers into percentages of the total number of respondents

#divide total number of respondents by each group
df_new= df_survey.div(df_survey.sum(1), axis=0)

#convert to %s and round percentages to 2d.p
df_new= round(100*df_new, 2)

df_new.head()
```

```python
colors_list=['#5cb85c', '#5bc0de','#d9534f']

ax= df_new.plot(kind= 'bar',
               width= 0.8,
               figsize= (20,8), 
               rot=90,
               color=colors_list,
               edgecolor= 'none'
              )

ax.set_title("Percentages of Respondent's Interest in Data Science areas",fontsize= 16)
ax.set_xlabel('Data Science Topics', fontsize=14)

plt.legend(labels=df_new.columns, loc='upper right', fontsize=14)


#remove the left, top, and right borders(spines)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)

# Only show ticks on the bottom spine
plt.xticks(fontsize=14) #or ax.xaxis.set_ticks_position('bottom')

#don't show ticks on the yaxis/left spine
plt.yticks([])

#to add annotations to plots, use ax i.e. Axes.annotate
#display the percentages above the bars
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height*1.02), ha='center', fontsize=14)
             
plt.show()
```
